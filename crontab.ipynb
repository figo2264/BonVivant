터미널 명령어
cron 리스트 확인
crontab -l

cron 시작
echo "30 8 * * 1-5 cd /Users/seunghakim/projects/Hanlyang_Stock && /Users/seunghakim/projects/Hanlyang_Stock/.venv/bin/python strategy.py >> /Users/seunghakim/projects/Hanlyang_Stock/cron.log 2>&1" | crontab -

cron 중지
crontab -r 또는 crontab -e # 편집기에서 원하는 줄을 삭제하고 저장

# 실시간 로그 확인
tail -f /Users/seunghakim/projects/Hanlyang_Stock/cron.log

# 최근 로그 확인
tail -20 /Users/seunghakim/projects/Hanlyang_Stock/cron.log
